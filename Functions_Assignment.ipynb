**Que. 6.Explain the concept of generators in python and how they are defined.**

**Ans.**
A Python generator function allows you to declare a function that behaves like an iterator, providing a faster and easier way to create iterators. They can be used on an abstract container of data to turn it into an iterable object like lists, dictionaries and strings.


It is fairly simple to create a generator in Python. It is as easy as defining a normal function, but with a yield statement instead of a return statement. If a function contains at least one yield statement (it may contain other yield or return statements), it becomes a generator function.

```python
def generatorFun():
    yield 1
    yield 2
    yield 3


for value in generatorFun():
    print(value)
```

**Que. 7.What are the advantages of using generators over regular functions?**

**Ans.**

**Memory efficiency:**
Generators only generate values when needed, so they don't need to store all values in memory at once. This makes them especially useful when working with large datasets or infinite sequences.

**Improved performance:**
Generators can be faster than regular loops because they avoid the overhead of creating a list or other data structure to store the results.

**Readability:**
Generators can make code easier to read and understand by breaking up the iteration process into smaller chunks.

**Composability:**
Generators can be easily composed to create new generators, enabling powerful data processing pipelines. This allows for modular and reusable code.

**Simplification:**
Generators simplify code for iterating over datasets without explicitly requiring state management.

**Que. 10.What is the difference between map, reduce(), and filter() functions in python?**

**Ans.**

**Map:**
Transforms each element in an array into a new array. For example, use of map to square the numbers in an array.

**Filter:**
Removes elements that don't meet a condition and creates a new array. For example, use filter to remove numbers less than 10 from an array.

**Reduce:**
Combines all elements in an array into a single value. For example, use reduce to add up all the numbers in an array.


```python
#map
def square(x):
  return x * x

numbers = [1, 2, 3, 4, 5]
squared = map(square, numbers)
print(list(squared))
```

```python
#filter
def is_even(x):
  return x % 2 == 0

numbers = [1, 2, 3, 4, 5]
even_numbers = filter(is_even, numbers)
print(list(even_numbers))
```

```python
#reduce
from functools import reduce

def add(x, y):
  return x + y

numbers = [1, 2, 3, 4, 5]
sum = reduce(add, numbers)
print(sum)
```

**Que. 4.Write a python function that checks if a given number is prime or not from 1 to 200.**

**Ans.**

```python
def is_prime(n):
    if n <= 1:
        return False
    for i in range(2, int(n ** 0.5) + 1):
        if n % i == 0:
            return False
    return True

for num in range(1, 201):
    print(f"{num} is {'a prime number' if is_prime(num) else 'not a prime number'}")
```

**Que. 5.Create an iterator class in python that generates the Fibonacci sequence up to a specified number of terms.**

**Ans.** This class implements the iterator protocol using the __iter__ and __next__ methods. The Fibonacci sequence is generated by keeping track of the current and next values in the sequence and updating them on each call to __next__.

```python
class FibonacciIterator:
    def __init__(self, num_terms):
        self.num_terms = num_terms
        self.current = 0
        self.next = 1
        self.index = 0

    def __iter__(self):
        return self

    def __next__(self):
        if self.index >= self.num_terms:
            raise StopIteration
        self.index += 1
        result = self.current
        self.current, self.next = self.next, self.current + self.next
        return result

fib_iterator = FibonacciIterator(10)
for number in fib_iterator:
    print(number)
```

**Que. 6.Write a generator function in python that yields the powers of 2 up to a given exponent.**

**Ans.**This function uses the yield keyword to generate the powers of 2 one at a time, up to the specified exponent

```python
def powers_of_two(exponent):
    for i in range(exponent + 1):
        yield 2 ** i

for power in powers_of_two(5):
    print(power)
```

**Que. 7.Implement a generator function that reads a file line by line and yields each line as a string.**

**Ans.**This function opens a file, iterates through each line, and yields it. The strip() method is used to remove any leading and trailing whitespace, including newlines.

```python
def read_lines(file_path):
    try:
        with open(file_path, 'r') as file:
            for line in file:
                yield line.strip()
    except FileNotFoundError:
        print(f"File not found: {file_path}")
    except IOError:
        print(f"Could not read file: {file_path}")

file_path = 'example.txt'
for line in read_lines(file_path):
    print(line)

```

**Que. 9.write a pyton program that uses 'map()' to convert a list of temperature from celsius to Fahrenheit.**

**Ans.**

```python
# Function to convert Celsius to Fahrenheit
def celsius_to_fahrenheit(celsius):
    return (celsius * 9/5) + 32

# List of temperatures in Celsius
celsius_temperatures = [0, 20, 37, 100]

# Convert to Fahrenheit using map
fahrenheit_temperatures = list(map(celsius_to_fahrenheit, celsius_temperatures))

print(fahrenheit_temperatures)
```

**Que. 10.Create a python program that uses 'filter()' to remove all the vowels from a given string.**

**Ans.**

```python
def remove_vowels(string):
    vowels = 'aeiouAEIOU'
    return ''.join(filter(lambda x: x not in vowels, string))

input_string = "Hello, World!"
result = remove_vowels(input_string)
print(result)
```
